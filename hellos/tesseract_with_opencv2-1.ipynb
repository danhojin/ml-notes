```python
# tesseret bounding box processing: https://stackoverflow.com/questions/20831612/getting-the-bounding-box-of-the-recognized-words-using-python-tesseract

def get_gray(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def get_RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
img = cv2.imread('./data/sample1.png')
plt.imshow(get_gray(img), cmap='gray')
```

```python
import pytesseract
from pytesseract import Output

d = pytesseract.image_to_data(img, output_type=Output.DICT)
n_boxes = len(d['level'])
for i in range(n_boxes):
    (x, y, w, h) = (d['left'][i], d['top'][i], d['width'][i], d['height'][i])
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

fig, ax = plt.subplots(figsize=(9, 6))
ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# url image: https://stackoverflow.com/questions/7391945/how-do-i-read-image-data-from-a-url-in-python
# tesseract usuage: https://www.learnopencv.com/deep-learning-based-text-recognition-ocr-using-tesseract-and-opencv/

import requests
from PIL import Image

url = 'https://www.learnopencv.com/wp-content/uploads/2018/06/computer-vision.jpg'
img = Image.open(requests.get(url, stream=True).raw)
```

```python
img
```

```python
# cv2_img = img.convert('BGR')

config = ('-l eng --oem 1 --psm 3')
text = pytesseract.image_to_string(img, config=config)
```
