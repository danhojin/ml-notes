\begin{equation}
f(x) \sim \mathcal{GP} (m(x), k(x, x'))
\end{equation}

for any finite subset $X = \{ x_1, \ldots x_n \}$ of the domain of x, the marginal distribution is a multivariate Gaussian distribution:

\begin{equation}
f(X) \sim \mathcal{N} (m(X), k(X, X))
\end{equation}

with mean vector $\mu = m(X)$ and covariance matrix $\Sigma = k(X, X)$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# Sampling from prior

### Brownian kernel

```python
points = 100
rfs = 5
X = np.linspace(0, 1, points)
# Brownian kernel
Σ = np.meshgrid(X, X)
Σ = np.minimum(*Σ)
```

```python
ys = np.random.multivariate_normal(
    mean=np.zeros(points),
    cov=Σ,
    size=rfs
)
```

```python
df = pd.DataFrame(ys.T, columns=[f'rf{i}' for i in range(rfs)])
df.index = X
df.plot()
```

### Squared exponential kernel

```python
points = 100
rfs = 5
X = np.linspace(0, 1, points)
# Brownian kernel
Σ = np.meshgrid(X, X)
Σ = np.exp(-100*(Σ[1] - Σ[0])**2)
```

```python
ys = np.random.multivariate_normal(
    mean=np.zeros(points),
    cov=Σ,
    size=rfs
)
```

```python
df = pd.DataFrame(ys.T, columns=[f'rf{i}' for i in range(rfs)])
df.index = X
df.plot()
```
