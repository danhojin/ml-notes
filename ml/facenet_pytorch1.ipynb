# Hello facenet-pytorch

Simple checkup for face detection flow with [facenet-pytorch](https://github.com/timesler/facenet-pytorch)

```python
from pathlib import Path
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image, ImageDraw

dshome = Path('./data/facenet')
img = Image.open(dshome / 'people3.jpg')
```

```python
img.size
```

```python
mtcnn = MTCNN()

# Detect faces
boxes, _ = mtcnn.detect(img)
boxes
```

```python
# Draw faces
img_draw = img.copy()
draw = ImageDraw.Draw(img_draw)
for box in boxes:
    draw.rectangle(box.tolist(), outline=(255, 0, 0), width=6)

img_draw.resize(map(lambda x: int(0.5*x), img.size))
```
